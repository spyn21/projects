```python
#importing the deendencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer
```

```python
#load Dataset to pandas DataFrame
credit_card_data=pd.read_csv('/content/drive/MyDrive/Dataset/creditcard.csv')
```

```python
#read first 5 rows
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# getting information about the data
credit_card_data.info()
```

```python
#check the number of missing value in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit and fraudal transaction
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalaced

0->Normal Transaction
1-> fraudal transaction

```python
#seperating data for analysis
Legit=credit_card_data[credit_card_data.Class==0]
Fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
Legit.Amount.describe()
```

```python
Fraud.Amount.describe()
```

```python
#compare the values of both transactions
credit_card_data.groupby('Class').mean()
```

Under sampling


building a sample dataset comntaining similar distribution of  normal transaction and fraudelant transaction

number of fraudal transaction -->492

```python
legit_sample=Legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,Fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting data into features and targets


```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

split the data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

since it is binary classification problem we use Logistic Regression Model


```python
model=LogisticRegression()
```

```python
#Traing The Logistic Regression Model with training Data
model.fit(x_train,y_train)
```

Model Evaluation based on accuracy_score


```python
#Accuracy on training data
x_train_prediction=model.predict(x_train)
Training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training data :',Training_data_accuracy)
```

```python
#Accuracy on test data
x_test_prediction=model.predict(x_test)
Test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on Test data :',Test_data_accuracy)
```
